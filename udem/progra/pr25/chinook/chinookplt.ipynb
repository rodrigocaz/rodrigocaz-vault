```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python

conn = sqlite3.connect('chinook.db')
query1 = "SELECT Country, COUNT(*) AS TotalClients FROM Customers GROUP BY Country ORDER BY TotalClients DESC;"
df1 = pd.read_sql_query(query1, conn)
print(df1.head())

plt.figure(figsize=(10,6))
plt.bar(df1['Country'], df1['TotalClients'], color='green')
plt.xlabel('Country')
plt.ylabel('Total Clients')
plt.title('Total Clients by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
query2 = 'SELECT Genres.Name AS SongGenre, COUNT(Tracks.TrackId) AS NumTracks FROM Tracks INNER JOIN Genres ON Tracks.GenreId = Genres.GenreId GROUP BY Genres.Name;'
df2 = pd.read_sql_query(query2, conn)
print(df2.head())

plt.figure(figsize=(12,8))
plt.bar(df2['SongGenre'], df2['NumTracks'], color='darkblue')
plt.xlabel('Song Genre')
plt.ylabel('Number of Tracks')
plt.title('Number of Tracks by Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
query3 = 'SELECT media_types.Name AS MediaType, COUNT(Tracks.TrackId) AS NumTracks FROM Tracks INNER JOIN media_types ON Tracks.MediaTypeId = media_types.MediaTypeId GROUP BY media_types.Name;'
df3 = pd.read_sql_query(query3, conn)
print(df3.head())

plt.figure(figsize=(12,8))
plt.barh(df3['MediaType'], df3['NumTracks'], color='darkred')
plt.xlabel('Media Type')
plt.ylabel('Number of Tracks')
plt.title('Number of Tracks by Media Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
