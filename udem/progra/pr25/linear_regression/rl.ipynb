La regresion lineal es un modelo de aprendizaje que se usa para que haya un relacion entre las variables x & y, x siendo la independiente & y, lo que hace es que hace una linea en la grafica y trata de minizmiar los errores entre los valores que son predecidos y los valores reales.

Overfitting: Esto pasa cuando los datos bien pero demsiado bien en el entrenamiento que a la hora de la prueba los datos no son los más buenos osea no tienen buen rendimiento

Underfitting: Esto pasa cuanda el modelo no ve el patron o conexion que hay entre las dos variables y eso lleva a un mal rendimiento del modelo, en el entrenamiento y en la prueba.

                        Interpretacion del Codigo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

Aqui se esta importando las librerias con las que hemos trabajado como pandas, numpy y matplotlib que son esenciales para el machine learning, pero aqui estamos impletendando seaborn que esta basada en matplotlib y nos permite usar graficos mas avanzados com graficos de distribucion que estaremos usando para comparar las predicciones y los resultados reales, ya por ultimo las librerias de sklearn las vamos a utlizar para dividir los modelos en prueba y entrenamiento, regresion linear y evualar el rendimiento de los modelos

```python
df = pd.read_csv('IowaHousingPrices-1.csv')
print(df.head())
```

Aqui se esta haciendo el dataframe con el pandas el cual esta leyendo el csv de los precios de las casas, y esta poniendo como output los heading de las casas

```python
x = df[['SquareFeet']].values
y = df[['SalePrice']].values
```

Aqui estamos definiendo las variables x & y con el dataframe de las variables del csv de los precios de las casas

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f'Tamaño del conjunto de entrenamiento: {x_train.shape}')
print(f'Tamaño del conjunto de prueba: {x_test.shape}')
```

Aqui estamos poniendo una variable conn varios argumentos los cuales son las variables de x & y que se van a entrenar y probar, se pone el output del tamaño del conjunto de entrenamiento y prueba

```python
modelo = LinearRegression()
modelo.fit(x_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intersección (b): {modelo.intercept_}")
```

Aqui estamos poniendo la variable modelo con la modulo de la libreria sklearn para hacer la regresion linear, y estamos poniendo como outputs los valores de la pendiente (m) e interseccion(b) del modelo los cuales nos sirven para la ecuacion en la grafica

```python
y_predict = modelo.predict(x_test)
mae = mean_absolute_error(y_test, y_predict)
mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)

print(f"Error Absoluto Medio (MAE): {mae}")
print(f"Error Cuadrático Medio (MSE): {mse}")
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse}")
```

Aqui estamos haciendo la variable de y_predict para con la variable modelo que tiene la regresion linear predecir el modelo con el arguemento de la prueba del modelo, hicimos variables para calcular el error absoluto medio y el error cuadratico medio que hay entre las pruebas y predicciones y otra variable que solo calcula la raiz de error cuadratico medio

```python
import matplotlib.pyplot as plt
plt.scatter(x_test, y_test, color='red', label='Actual')
plt.plot(x_test, y_predict, color='black', label = 'Predicted')
plt.xlabel('Square Feet')
plt.ylabel('Sale Price')
plt.legend()
plt.show()
```

Aqui ya solo estamos poniendo los valores de los modelos y la regresion lineal, osea estamos poniendo los valores del modelo actual como quedaron y pues estan esparcidos por el mapa pero muchos cerca de lo predecido lo cual es la linea negra en la grafica con sus respectivas variables x & y.
